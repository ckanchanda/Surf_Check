```python
#Open CSV
final_beach_df = pd.read_csv("../Resources/beach_data.csv")
final_beach_df
```

```python
# Renamed "Unnamed: 0" column and set it as the index
final_beach_df = final_beach_df.rename(columns={'Unnamed: 0': ''})
final_beach_df = final_beach_df.set_index('beach_id')
final_beach_df
```

```python
### BY WAVE HEIGHT
```

```python
#New DF for wave height
wave_height_df = final_beach_df[["Beach_Name", "Wave_Height"]].copy()
wave_height_df
```

```python
wave_height_df.describe()
```

```python
# Flatten column MultiIndex
wave_height_df.columns = [x[0] for x in wave_height_df.columns]
```

```python
wave_height_bins = [0, 3, 9, 12]
wave_sizes = ["Small (<4)", "Medium (4-9)", "Large (10-12)"]


# Categorize based on the bins.
final_beach_df["Wave Size"] = pd.cut(final_beach_df["Wave_Height"], wave_height_bins, labels=wave_sizes)
final_beach_df
```
